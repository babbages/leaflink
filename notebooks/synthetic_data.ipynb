```python
import pandas as pd
import numpy as np

np.random.seed(42)  # Set a seed for reproducibility

hours = 3 * 7 * 24  # 3 weeks of data with hourly intervals
dates = pd.date_range('20220101', periods=hours, freq='H')  # Create a datetime index

# Generate synthetic data using random numbers
temperature_c = np.random.normal(20, 1, hours)
humidity = np.random.normal(50, 3, hours)

# Generate soil moisture data based on humidity
moisture_multiplier = 50
soil_moisture = 20000 + (humidity - humidity.min()) * moisture_multiplier
trend_factor = np.linspace(-5000, 5000, hours)
soil_moisture = np.clip(soil_moisture + trend_factor, 20000, 22000)

# Depend light intensity on temperature and humidity
temperature_threshold = 22
humidity_threshold = 40
light_intensity = np.where((temperature_c > temperature_threshold) & (humidity < humidity_threshold), 1, 0)

temperature_f = temperature_c * 9/5 + 32

# Create a pandas DataFrame
data = pd.DataFrame({
    'date_time': dates,
    'temperature_c': temperature_c,
    "temperature_f": temperature_f,
    'humidity': humidity,
    'soil_moisture (%)': soil_moisture,
    'light': light_intensity
}) # Display the first 5 rows
# data.to_csv('synthetic_plant_sensor_data.csv', index=False)  # Save the data to a CSV file
```
